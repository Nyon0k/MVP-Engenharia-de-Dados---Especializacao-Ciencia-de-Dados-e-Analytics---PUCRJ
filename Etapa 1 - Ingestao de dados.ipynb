```python
import pandas as pd
import requests
```

```python
# Link compartilhado
url = "https://drive.google.com/file/d/14yuZrUKYQiGhceAF4rYOoQF5G2Mgxx2A/view?usp=sharing"
file_id = url.split("/d/")[1].split("/")[0]
download_url = f"https://drive.google.com/uc?id={file_id}"

# Caminho dentro do seu volume
volume_path = "/Volumes/puc/bronze/data/base_pd.parquet"

print("Baixando arquivo...")
r = requests.get(download_url)
r.raise_for_status()

# Criar arquivo diretamente no volume
with open(volume_path, "wb") as f:
    f.write(r.content)

print("Arquivo salvo no Volume com sucesso!")
```

```python
src = "/Volumes/puc/bronze/data/base_pd.parquet"
dst = "/Volumes/puc/bronze/data/base_pd_fixed.parquet"

df = pd.read_parquet(src)

# Converte timestamps nanos â†’ micros
for col in df.select_dtypes(include=["datetime64[ns]"]):
    df[col] = df[col].astype("datetime64[us]")

df.to_parquet(dst, index=False)

print("Arquivo salvo:", dst)
```
